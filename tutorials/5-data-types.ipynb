## Python Data Types

### 1. **Numeric Types:**
   - **int:** Integer type represents whole numbers without any decimal point.
     ```python
     x = 10
     ```

   - **float:** Float type represents floating-point numbers with decimal points.
     ```python
     y = 3.14
     ```

   - **complex:** Complex type represents complex numbers in the form of `real + imagj`.
     ```python
     z = 2 + 3j
     ```

### 2. **Text Type:**
   - **str:** String type represents sequences of characters enclosed in single or double quotes.
     ```python
     name = "Python"
     ```

### 3. **Sequence Types:**
   - **list:** List type represents ordered and mutable sequences.
     ```python
     numbers = [1, 2, 3, 4]
     ```

   - **tuple:** Tuple type represents ordered and immutable sequences.
     ```python
     coordinates = (5, 6)
     ```

   - **range:** Range type represents a sequence of numbers within a specified range.
     ```python
     nums = range(0, 5)
     ```

### 4. **Mapping Type:**
   - **dict:** Dictionary type represents key-value pairs enclosed in curly braces.
     ```python
     person = {'name': 'John', 'age': 30}
     ```

### 5. **Set Types:**
   - **set:** Set type represents an unordered collection of unique elements.
     ```python
     unique_numbers = {1, 2, 3, 4}
     ```

   - **frozenset:** Frozenset type represents an immutable set.
     ```python
     frozen_set = frozenset([1, 2, 3])
     ```

### 6. **Boolean Type:**
   - **bool:** Boolean type represents either `True` or `False`.
     ```python
     is_python_fun = True
     ```

### 7. **Binary Types:**
   - **bytes:** Bytes type represents a sequence of bytes.
     ```python
     binary_data = b'hello'
     ```

   - **bytearray:** Bytearray type represents a mutable sequence of bytes.
     ```python
     mutable_binary_data = bytearray(b'world')
     ```

   - **memoryview:** Memoryview type provides a view of an object's memory.
     ```python
     view = memoryview(b'example')
     ```

### 8. **None Type:**
   - **NoneType:** None type represents the absence of a value or a null value.
     ```python
     no_value = None
     ```

Understanding and using these data types is crucial for effective programming in Python.
